```python
# 3. Create a dataset representing categories and their corresponding values. Compare different 
# categories based on numerical values.
```

```python
data = {
    'Category' : ['A', 'B', 'C', 'D', 'E'],
    'Value' : [25, 40, 30, 35, 20]
}

df = pd.DataFrame(data)
```

```python
sns.set(style = 'whitegrid')

plt.figure(figsize = (10, 6))
sns.barplot(x = 'Category', y = 'Value', data = df, palette = 'viridis')

plt.title('Comparison of Different Categories')
plt.xlabel('Category')
plt.ylabel('Values')

plt.show()
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a 
# numerical variable across different categories.
```

```python
np.random.seed(0)

categories = ['A', 'B', 'C', 'D', 'E']
n = 200

data = {'Category' : np.random.choice(categories, size = n),
        'Value' : np.random.normal(0, 1, size = n)
       }

category_means = {
    'A' : 10, 'B' : 20, 'C' : 15, 'D' : 25, 'E' : 5
}

df = pd.DataFrame(data)
df["Value"] += df["Category"].map(category_means)

sns.set(style = 'whitegrid')

plt.figure(figsize = (10,6))
sns.boxplot(x = 'Category', y = 'Value', data = df, palette = 'viridis')

plt.title('Distribution of Values of Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a 
# dataset using a heatmap.
```

```python
np.random.seed(42)
n = 100

x1 = np.random.normal(0, 1, n)
x2 = 2.5 * x1 + np.random.normal(0, 1, n)
x3 = -1.5 * x1 + np.random.normal(0, 1, n)
x4 = np.random.normal(0, 1, n)

df = pd.DataFrame({'Feature1' : x1, 
                  'Feature2' : x2,
                  'Feature3' : x3,
                  'Feature4' : x4})

print(df.head())
```

```python
corr_matrix = df.corr()
print(corr_matrix)
```

```python
sns.set(style = 'white')

plt.figure(figsize = (8,6))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', square = True, linewidths = .5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
